```python
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('./datajam-2021/train.csv')
test_data=pd.read_csv('./datajam-2021/test.csv')
# train_data.shape
```

```python
#train_data.describe()
```

```python
train_data.drop(train_data.tail(2).index, inplace = True)
train_data
```

```python
X= train_data.drop(columns=['motor_UPDRS'])
Y= train_data['motor_UPDRS']
```

```python
# X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.2)
```

```python
model= KNeighborsRegressor(n_neighbors = 1)
model.fit(X,Y)
```

```python
predictions = model.predict(test_data)
predictions
```

```python
predictions.size
```

```python
# initialize data of lists
submission=pd.DataFrame({'Id':test_data['Id'],'motor_UPDRS':predictions})



#Print the output
submission.head()
```
